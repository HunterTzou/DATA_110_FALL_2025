#Statistical Summary

For numerical data, `df.describe()` provides key statistics:

- count: number of non-null values
- mean: average value
- std: standard deviation
- min: minimum value
- 25%, 50%, 75%: quartile values
- max: maximum value
- This helps you understand the distribution of your numerical data.

```python
df.describe()
```
