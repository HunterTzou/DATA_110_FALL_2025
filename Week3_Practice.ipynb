```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = ['a','b','c','d']
y = [10,8,12,16]
z = [11,7,10,9]

df = pd.read_csv('https://raw.githubusercontent.com/Zhouhun/Data110-21843--fall25/refs/heads/main/dataset/happiness.csv')
df.head()
```

```python
# Sort the DataFrame by the "Happiness Score" column (ascending by default)
df_sorted = df.sort_values(by='Happiness Score')

# Display the first 5 rows to quickly check if sorting worked
df_sorted.head()

# Create a new figure for the plot and set its size (width=12, height=8)
plt.figure(figsize=(12, 8))

# Create a horizontal bar chart:
# y-axis = countries, x-axis = their happiness scores
plt.barh(df_sorted['country'], df_sorted['Happiness Score'])

# Label the x-axis for clarity
plt.xlabel('Happiness Score')

# Label the y-axis to show what the categories represent
plt.ylabel('Country')

# Add a title so the plot is self-explanatory
plt.title('Happiness Score by Country')

# Show the plot
plt.show()
```
